## Linear Search

```python
arr = []
print('Enter 5 array elements:')
for i in range(5):
    arr.append(int(input()))

target = 20
found = False

for i in range(len(arr)):
    if arr[i] == target:
        print('Element found at index:', i)
        found = True
        break

if not found:
    print('Element not found in list!')
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 100)

linear = n                # O(n)
binary = np.log2(n)       # O(log n)

plt.figure()
plt.plot(n, linear)
plt.plot(n, binary)

plt.xlabel("Input Size (n)")
plt.ylabel("Number of Operations")
plt.title("Linear Search vs Binary Search")

plt.legend(["Linear Search O(n)", "Binary Search O(log n)"])
plt.show()
```

## Binary Search

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 3, 5, 7, 9, 11]
result = binary_search(arr, 7)
print('Element found at index:' if result != -1 else 'Element not found', result)
```

## Bubble Sort

```python
arr = [9, 5, 1, 4, 3]
n = len(arr)

for i in range(n-1):
    swapped = False
    for j in range(n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            swapped = True
    if not swapped:
        break

print('Sorted Array:', arr)
```

## Selection Sort

```python
arr = [64, 25, 12, 22, 11]
n = len(arr)

for i in range(n):
    min_index = i
    for j in range(i+1, n):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

print('Sorted Array:', arr)
```

## Insertion Sort

```python
arr = [12, 11, 13, 5, 6]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key

print('Sorted Array:', arr)
```

## Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort(arr)
print('Sorted Array:', arr)
```

# Factorial (Iterative & Recursive)

```python
# Factorial - Iterative
def factorial_iterative(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact


# Factorial - Recursive
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)


num = 5
print("Factorial (Iterative):", factorial_iterative(num))
print("Factorial (Recursive):", factorial_recursive(num))
```

# Fibonacci (Iterative & Recursive with Step Count)

```python
# Fibonacci with Step Count

step_count = 0

# Iterative Fibonacci
def fib_iterative(n):
    global step_count
    step_count = 0
    
    a, b = 0, 1
    print("Fibonacci Series:", a, b, end=" ")
    
    for i in range(2, n):
        c = a + b
        print(c, end=" ")
        a, b = b, c
        step_count += 1
        
    print("\nStep Count (Iterative):", step_count)


# Recursive Fibonacci
def fib_recursive(n):
    global step_count
    step_count += 1
    
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)


n = 6
fib_iterative(n)

step_count = 0
print("Recursive Fibonacci:", fib_recursive(n))
print("Step Count (Recursive):", step_count)
```
